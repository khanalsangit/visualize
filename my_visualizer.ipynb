```python
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
def visualizer(x, y):
    visualize = plt.plot(x, y)
    return visualize
```

```python
df=pd.read_csv('student_data.csv')
df.head()
```

```python
x_data = df['day']
y_data = df['no_of_students']
visualizer(x_data, y_data)
```

```python
with open('visualizer_function.pkl', 'wb') as file:
    pickle.dump(visualizer, file)
```
